## Text Example

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pymc as mc
%matplotlib inline

baseDir = r"C:\Users\tfolkman\Desktop\TFolkman\Bayes for Hackers\Data"

count_data = np.loadtxt(baseDir + "/txtdata.csv")
n_count_data = len(count_data)
plt.bar(np.arange(n_count_data), count_data, color="#348ABD")
plt.xlabel("Time (days)")
plt.ylabel("count of text-msgs received")
plt.title("Did the user's texting habits change over time?")
plt.xlim(0, n_count_data);
```

Here we have count data. A poisson distribution is good for count data, so we could say that Counts ~ Poisson($\lambda$). I don't know what $\lambda$ is, but it looks like it might increase later on in the period becasue the counts increase.

Thus I need prior distributions for $\lambda$1 and $\lambda$2. $\lambda$ can be any positive number, so the exponential distribution might be a good choice to model $\lambda$. We can say $\lambda$ ~ Exponentail($\tau$). A good rule of thumb for choosing $\tau$ is the inverse of the average of the count data.

```python

colours = ["#348ABD", "#A60628"]
a = np.linspace(0, 4, 100)
expo = stats.expon
lambda_ = [0.5, 1]

for l, c in zip(lambda_, colours):
    plt.plot(a, expo.pdf(a, scale=1./l), lw=3,
             color=c, label="$\lambda = %.1f$" % l)
    plt.fill_between(a, expo.pdf(a, scale=1./l), color=c, alpha=.33)

plt.legend()
plt.ylabel("PDF at $z$")
plt.xlabel("$z$")
plt.title("Probability density function of an Exponential random variable;\
 differing $\lambda$")
```

Above is the PDF of the exponential function. A PDF assigns probabilities to the different outcomes that can take place

But how do we know at what day $\lambda$ switches? This is hard to pick because of the noiseness of the data. 

```python

alpha = 1.0/count_data.mean()  # Recall count_data is the
                               # variable that holds our txt counts
lambda_1 = mc.Exponential("lambda_1", alpha)
lambda_2 = mc.Exponential("lambda_2", alpha)

tau = mc.DiscreteUniform("tau", lower=0, upper=n_count_data)
```

Here $\lambda_{1}$, $\lambda_{2}$, and $\tau$ are defined as stochastic variables - treated as random number generators. Tau is basically a uniform distribution across all days - hence each day is equally probable

```python
samples = [lambda_1.random() for i in range(20000)]
plt.hist(samples, bins=70, normed=True, histtype="stepfilled")
plt.title("Prior distribution for $\lambda_1$")
```

This is like P(A) in Bayes therom

```python
@mc.deterministic
def lambda_(tau=tau, lambda_1=lambda_1, lambda_2=lambda_2):
    out = np.zeros(n_count_data)
    out[:tau] = lambda_1  # lambda before tau is lambda1
    out[tau:] = lambda_2  # lambda after (and including) tau is lambda2
    return out
```

The lambda function returns a random array. It randomly draws $\tau$ to determine how many days - starting from the beginning - should have $\lambda_{1}$, and the rest are given $\lambda_{2}$ - both of which are random variables

```python
observation = mc.Poisson("obs", lambda_, value=count_data, observed=True)
model = mc.Model([observation, lambda_1, lambda_2, tau])
```

The above basically allows me to get the P(Data|$\lambda_{1}$,$\lambda_{2}$,$\tau$). It first can sample $\lambda_{1}$,$\lambda_{2}$ and $\tau$ and then can sample counts conditional on these values. The model kind of puts everything together.

```python
mcmc = mc.MCMC(model)
mcmc.sample(40000, 10000, 1)

lambda_1_samples = mcmc.trace('lambda_1')[:]
lambda_2_samples = mcmc.trace('lambda_2')[:]
tau_samples = mcmc.trace('tau')[:]

ax = plt.subplot(311)
ax.set_autoscaley_on(False)

plt.hist(lambda_1_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\lambda_1$", color="#A60628", normed=True)
plt.legend(loc="upper left")
plt.title(r"""Posterior distributions of the variables
    $\lambda_1,\;\lambda_2,\;\tau$""")
plt.xlim([15, 30])
plt.xlabel("$\lambda_1$ value")

ax = plt.subplot(312)
ax.set_autoscaley_on(False)
plt.hist(lambda_2_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\lambda_2$", color="#7A68A6", normed=True)
plt.legend(loc="upper left")
plt.xlim([15, 30])
plt.xlabel("$\lambda_2$ value")

plt.subplot(313)
w = 1.0 / tau_samples.shape[0] * np.ones_like(tau_samples)
plt.hist(tau_samples, bins=n_count_data, alpha=1,
         label=r"posterior of $\tau$",
         color="#467821", weights=w, rwidth=2.)
plt.xticks(np.arange(n_count_data))

plt.legend(loc="upper left")
plt.ylim([0, .75])
plt.xlim([35, len(count_data)-20])
plt.xlabel(r"$\tau$ (in days)")
plt.ylabel("probability");
```

## A Simple Case

Imagine we are interested in estimating the true probability of buyers who go to Site A purchasing. To do this in a Bayesian way, we first need to assign a prior distribution. Say we really have no idea, so we can do the uniform distribution

```python
p = mc.Uniform('p',lower=0,upper=1)
```

Now imagine we have some data on 1500 obs with a true 5% chance of a purchase (in reality this is unknown)

```python
p_true = .05
N = 10000
occurrences = mc.rbernoulli(p_true,N)
print occurrences.sum()
```

The observed frequency is:

```python
print "What is the observed frequency in Group A? %.4f" % occurrences.mean()
print "Does this equal the true frequency? %s" % (occurrences.mean() == p_true)
```

We can combine this knowledge (data and our prior belief of the distribution) and run an inference algo

```python
obs = mc.Bernoulli("obs",p,value=occurrences,observed=True)
mcmc = mc.MCMC([p, obs])
mcmc.sample(18000, 1000)

plt.title("Posterior distribution of $p_A$, the true effectiveness of site A")
plt.vlines(p_true, 0, 90, linestyle="--", label="true $p_A$ (unknown)")
plt.hist(mcmc.trace("p")[:], bins=25, histtype="stepfilled", normed=True)
plt.legend()
```

As we increase the value of N the spread of the distribution decreases

Imagine we are intereted in knowing the difference between P(Purchase if see site A) and P(Purchase if see site B):

```python
#unknown quantities
true_p_A = .05
true_p_B = .04

#Sample Sizes
N_A = 1500
N_B = 750

#generate data
observations_A = mc.rbernoulli(true_p_A,N_A)
observations_B = mc.rbernoulli(true_p_B,N_B)

print "Observed P(A): {}".format(observations_A.mean())
print "Observed P(B): {}".format(observations_B.mean())
```

```python
#Set up priors - assume uniform (i.e. have no idea probability)
p_A = mc.Uniform("p_A",0,1)
p_B = mc.Uniform("p_B",0,1)

#Define deterministic function to get at unknown of interest
@mc.deterministic
def delta(p_A=p_A, p_B=p_B):
    return p_A - p_B

# Set of observations, in this case we have two observation datasets.
obs_A = mc.Bernoulli("obs_A", p_A, value=observations_A, observed=True)
obs_B = mc.Bernoulli("obs_B", p_B, value=observations_B, observed=True)

mcmc = mc.MCMC([p_A, p_B, delta, obs_A, obs_B])
mcmc.sample(20000, 1000)

p_A_samples = mcmc.trace("p_A")[:]
p_B_samples = mcmc.trace("p_B")[:]
delta_samples = mcmc.trace("delta")[:]

#histogram of posteriors

ax = plt.subplot(311)

plt.xlim(0, .1)
plt.hist(p_A_samples, histtype='stepfilled', bins=25, alpha=0.85,
         label="posterior of $p_A$", color="#A60628", normed=True)
plt.vlines(true_p_A, 0, 80, linestyle="--", label="true $p_A$ (unknown)")
plt.legend(loc="upper right")
plt.title("Posterior distributions of $p_A$, $p_B$, and delta unknowns")

ax = plt.subplot(312)

plt.xlim(0, .1)
plt.hist(p_B_samples, histtype='stepfilled', bins=25, alpha=0.85,
         label="posterior of $p_B$", color="#467821", normed=True)
plt.vlines(true_p_B, 0, 80, linestyle="--", label="true $p_B$ (unknown)")
plt.legend(loc="upper right")

ax = plt.subplot(313)
plt.hist(delta_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of delta", color="#7A68A6", normed=True)
plt.vlines(true_p_A - true_p_B, 0, 60, linestyle="--",
           label="true delta (unknown)")
plt.vlines(0, 0, 60, color="black", alpha=0.2)
plt.legend(loc="upper right")

# Count the number of samples less than 0, i.e. the area under the curve
# before 0, represent the probability that site A is worse than site B.
print "Probability site A is WORSE than site B: %.3f" % \
    (delta_samples < 0).mean()

print "Probability site A is BETTER than site B: %.3f" % \
    (delta_samples > 0).mean()
```
