# ANOMOLY DETECTION

Read in data

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv(r"\\laproj\Public\tfolkman\BizDev\Class Action\Firm Data\Gibson Dunn\WebPull.csv",
                   parse_dates=['Date'],index_col='Date')
#Set Count Data
data['Count'] = 1

#Subset
data = data['2011':]
```

Look at monthly counts

```python
t = data.resample('M',how='sum')
t.plot()
plt.title("Number of Cases for Covington Burling\nMonthly")
```

Look at Poisson Distribution With $\lambda$ equal to mean

```python
mean = t['Count'].mean()
poi = stats.poisson
a = range(0,200)

plt.bar(a, poi.pmf(a, mean), color="#348ABD",
        label="$\lambda = %.1f$" % mean, alpha=0.60,
        edgecolor="#348ABD", lw="3")
```

Not going to work because the outlier shifts the mean too much

Lets look at clustering:

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(t)
t['Cluster'] = kmeans.predict(t)
test = t.groupby(t['Cluster'])

colours = ["#348ABD", "#A60628"]
i = 0
for name,group in test:
    plt.bar(group.index, group['Count'],color=colours[i],edgecolor=colours[i],lw=3,width=.25)
    i+=1
```
